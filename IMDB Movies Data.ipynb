```python
import numpy as np
import pandas as pd
```

```python
#read Movies.csv file
movies=pd.read_csv('movies.csv',index_col=0)
movies.head()
```

```python
#read director.csv file
directors=pd.read_csv('directors.csv',index_col=0)
directors.head()
```

```python
#merge movies and directors dataset for Analysis
df=movies.merge(directors,how='left',left_on='director_id',right_on='id')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include=object)
```

```python
#remove duplicate columns director_id and id_y
df.drop(['id_y','director_id'],axis=1,inplace=True)
df.head()
```

```python
movies['director_id'].nunique()
```

```python
directors['id'].nunique()
```

```python
#check if all director_id values are present in id
np.all(movies['director_id'].isin(directors['id']))
```

```python
#convert the values of revenue into million dollars USD
df['revenue'] = (df['revenue']/1000000).round(2)
df.head()
```

```python
#convert the values of budget into million dollars USD
df['budget'] = (df['budget']/1000000).round(2)
df.head()
```

```python
#highly rated movies
#movies with ratings > 7
df.loc[df['vote_average']>7]
```

```python
df.loc[df['vote_average']>7,['title','director_name']]
```

```python
#filter highly rated movies released after 2014
df[(df['vote_average']>7) & (df['year']>2014)]
```

```python
#movies released on either Friday or Sunday
df[(df['day']=='Friday') | (df['day']=='Sunday')]
```

```python
#Top 5 most popular movies
df.sort_values(['popularity'],ascending=False).head()
```

```python
#list of movies directed by a particular director, say 'Christopher Nolan'
df.loc[df['director_name']=='Christopher Nolan',['title']]
```

```python
#monthwise count of movies released
month_wise_count= df.groupby('month')['title'].count().reset_index()
month_wise_count.columns=(['Months','Count'])
month_wise_count.sort_values(by='Count',ascending=False)
```
